## Cursors

This will help us return > 100 results for our searches!

```python
# cursor in twython -- cursors are relevant when you are "paging" through results, tracking your place
# .cursor() takes another function as a parameter
# then we put all the parameters we would have passed to our twitter.search function

cursor = twitter.cursor(twitter.search, q='"data journalism" -filter:retweets', count=100)
all_text = list()

for tweet in cursor:
    all_text.append(tweet['text'])
    if len(all_text) > 500: # twitter has rate limits, so we're building in a limit to our request 
        break # this aborts the loop
```

```python
all_text
```

```python
from collections import Counter
```

```python
# create a counter object
c = Counter()
```

```python
for item in all_text:
    c.update(item.split())

c.most_common(25)
```

**Stopwords** are words that are common across all texts, so you can safely exclude them (words like "the", "I", "you", etc.). Usually stopwords are shorter than other words. 

```python
c = Counter()
for item in all_text:
    c.update([w for w in item.split() if len(w) > 3])

c.most_common(25)
```

```python
c = Counter()
for item in all_text:
    # to fold both lowercase and uppercase versions as one word 
    c.update([w.lower() for w in item.split() if len(w) > 3])

c.most_common(25)
```

```python
cursor = twitter.cursor(twitter.search, q='"data journalism" -filter:retweets', count=100)
all_media_urls = list()

for tweet in cursor:
    if 'media' in tweet['entities']:
        for item in tweet['entities']['media']:
            all_media_urls.append(item['media_url'])
    if len(all_media_urls) > 1000: 
        break 
```

```python
all_media_urls
```

```python
fh = open("preview.html", "w")
for item in all_media_urls:
    fh.write('<img src="{}" width="50">'.format(item))
fh.close()
```

```python
import pg8000
```

```python
lakes = []

conn = pg8000.connect(database='mondial')
cursor = conn.cursor()
cursor.execute('SELECT name, area, depth, elevation, type, river FROM lake')
for row in cursor.fetchall():
    lake = {'name': row[0],
           'area': row[1],
           'depth': row[2],
           'elevation': row[3],
           'type': row[4],
           'river': row[5]}
    lakes.append(lake)
len(lakes)
```

```python
import json
json.dump(lake, open(lakes))
```

```python
#to generate a fact sentence
#pick a column from the record
#compose a sentence based on that column
#from a pre-existing list of templates
```

```python
#Templates:
# The area of {} is {} square kilometers.
# The depth of {} is {} meters.
#etc.

#Process:
#randomly select a column
#if that column is not empty
# ...fill in the template with the lake name and the value for that column
```

```python
sentences = {
    'area': 'The area of {} is {} square kilometers.',
    'depth': 'The depth of {} is {} meters.',
    'elevation': 'The elevation of {} is {} meters.',
    'type': 'The type of {} is "{}."',
    'river': '{} empties into a river named {}.'
}
```

```python
"having a string like {} and {}".format("aardvark", "beluga")
```

```python
import random
```

```python
lake = random.choice(lakes)
col = random.choice(list(lake.keys()))
print(lake)
print(col)
sentence_template = sentences[col]
sentence_template.format(lake['name'], lake[col])
```

```python
twitter.update_status(status=sentence_template.format(lake['name'], lake[col]))
```

```python
import random
def random_lake_sentence(lakes, sentences):
    rlake = random.choice(lakes)
    # get the keys in the dictionary whose value is not None; we'll only try to
    # make sentences for these
    possible_keys = [k for k, v in rlake.items() if v is not None and k != 'name']
    rframe = random.choice(possible_keys)
    return sentences[rframe].format(rlake['name'], rlake[rframe])
for i in range(10):
    print(random_lake_sentence(lakes, sentences))
```

```python
twitter.update_status(status=random_lake_sentence(lakes, sentences))
```

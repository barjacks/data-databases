```python
cursor = twitter.cursor(twitter.search, q='"data jouralism" -filter:retweets', count=100)
all_text = list()
for tweet in cursor['statuses']:
    all_text.append(tweet['text'])
    if len(all_text) > 500:
        break

        
        
```

```python
from collections import Counter
c = Counter()
for item in all_text:
    c.update([w.lower() for w in item.split() if len(w) >3])
print(c.most_common(25))
```

# entities

```python
response = twitter.search(q="data journalism", result_type="recent", count=20)
```

```python
cursor = twitter.cursor(twitter.search, q='"kevin durant" -filter:retweets', count=100)
all_urls = list()
for tweet in cursor:
    for item in tweet['entities']['urls']:
        all_urls.append(item['expanded_url'])
    if len(all_urls) > 1000:
        break
```

```python
url_counter = Counter(all_urls)
for item in url_counter.most_common(10):
    print(item)
```

```python
cursor = twitter.cursor(twitter.search, q='"data journalism" -filter:retweets', count=100)
all_media_urls = list()
for tweet in cursor:
    if 'media' in tweet['entities']:
        for item in tweet['entities']['media']:
            all_media_urls.append(item['media_url'])
    if len(all_urls) > 1000:
        break
```

```python
all_media_urls
```

```python
fh = open('preview.html', 'w')
for item in all_media_urls:
    fh.write('<img src_"{} width="50">'.format(item))
fh.close()
```
